```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import quantile_transform

microarray_data = pd.read_csv('/kaggle/input/gse62232-log2-quantile-normalization/GSE622322.csv')
microarray_data.set_index("ID_REF", inplace=True)
log2_transformed_data = np.log2(microarray_data)

quantile_normalized_data = quantile_transform(log2_transformed_data, n_quantiles=50, axis=0, random_state=0, copy=True)

quantile_normalized_data = pd.DataFrame(quantile_normalized_data, columns=microarray_data.columns)
quantile_normalized_data.to_csv('GSE62232_log_quantile_normalized_data.csv')
```
